```python
# Re-import libraries due to code execution state reset
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Step 1: Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

# Step 2: Sample data
x1 = 3
x2 = 2
y_actual = 17

# Step 3: Initialize weights
np.random.seed(42)
w1 = np.random.uniform(1, 10)
w2 = np.random.uniform(1, 10)

# Step 6: Training parameters
learning_rate = 0.01
epochs = 20

# To store progression for plotting
errors = []
ypreds = []

# Training loop
for epoch in range(epochs):
    # Feed Forward
    y_pred = x1 * w1 + x2 * w2
    
    # Error calculation
    error = (y_actual - y_pred) ** 2
    
    # Store for visualization
    errors.append(error)
    ypreds.append(y_pred)
    
    # Backpropagation (manual gradient descent)
    derror_dypred = -2 * (y_actual - y_pred)
    dypred_dw1 = x1
    dypred_dw2 = x2
    
    # Gradient for weights
    dw1 = derror_dypred * dypred_dw1
    dw2 = derror_dypred * dypred_dw2
    
    # Update weights
    w1 = w1 - learning_rate * dw1
    w2 = w2 - learning_rate * dw2

# Plotting the error and predictions
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

ax1.plot(errors, marker='o')
ax1.set_title("Error Across Epochs")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Squared Error")

ax2.plot(ypreds, marker='x', color='green')
ax2.set_title("Predicted y Across Epochs")
ax2.set_xlabel("Epoch")
ax2.set_ylabel("Predicted y")

plt.tight_layout()
plt.show()

# Final predictions and weights
w1, w2, y_pred, error
```
